## Analysis using NLP libraries

```python
import pandas as pd
import spacy
import nltk
import en_core_web_sm
import re 
nlp = en_core_web_sm.load() #initializing stopwords
stop_words = spacy.lang.en.stop_words.STOP_WORDS
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from nltk.stem.porter import PorterStemmer
from collections import Counter

stemmer = PorterStemmer()
```

```python
# Create function to clean the data

def remove_content(text):
    text = re.sub(r"http\S+", "", text) #remove urls
    text=re.sub(r'\S+\.com\S+','',text) #remove urls
    text=re.sub(r'\@\w+','',text) #remove mentions
    text =re.sub(r'\#\w+','',text) #remove hashtags
    return text
def process_text(text, stem=False): #clean text
    text=remove_content(text)
    text = re.sub('[^A-Za-z]', ' ', text.lower()) #remove non-alphabets
    tokenized_text = word_tokenize(text) #tokenize
    clean_text = [
         word for word in tokenized_text
         if (word not in stop_words and len(word)>1)
    ]
    if stem:
        clean_text=[stemmer.stem(word) for word in clean_text]
    return ' '.join(clean_text)

def extract_ngrams(data, num):
    n_grams = ngrams(nltk.word_tokenize(data), num)

    n_grams_ = [grams for grams in n_grams]

    return n_grams_
```

#### Get most word written in tweet
    word in one gram
    word in two or bigram

1. One gram word most mention

```python
# 1. 1gram

df_sprm_['cleaned_tweets']=df_sprm_['tweet'].apply(lambda x: process_text(x))

tweet_list = df_sprm_['cleaned_tweets'].tolist()

# Keywords 1gram

one_gram =' '.join(tweet_list).split()
aa = Counter(one_gram)

print(aa.most_common(30))
list_1gram = aa.most_common(30)
```

We can analyse the sentiment from text. Text(tweet) mostly written in Bahasa Malaysia, we use this library and pretrained model.

```python
%%time
import malaya
```

```python
model = malaya.sentiment.multinomial()
```

```python
# we analyse this list of tweet
# for display purpose only
tweet_list[1:10]
```

```python
# output in dict
output_ = model.predict_proba(tweet_list)
output_
```

```python
print(len(output_))
```

We get average sentiment for positive,negative and neutral

```python
# Average Positive

total_positive = []

for i in output_ :

    value_positif = i.get('positive')
    total_positive.append(value_positif)

#print(total_positif)
average_positive = sum(total_positive)/519
print(average_positive)
```

```python
# Average Negative

total_negative = []

for i in output_ :

    value_negative = i.get('negative')
    total_negative.append(value_negative)

#print(total_negative)
average_negative = sum(total_negative)/519
print(average_negative)
```

```python
# calculate average

total_neutral = []

for i in output_ :

    value_neutral = i.get('neutral')
    total_neutral.append(value_neutral)

#print(total_negative)
average_neutral = sum(total_neutral)/519
print(average_neutral)
```

```python
list_sentiment = []

for i in output_:

    max_key = max(i, key = i.get)
    list_sentiment.append(max_key)

list_sentiment[0:3]
```

```python
# create column to store sentiment

df_sprm_['predict'] = list_sentiment
```

```python
list_score = []

for i in output_:

    values_score = i.values()
    max_value = max(values_score)

    list_score.append(max_value)

list_score[0:3]
```

```python
# create column to store sentiment

df_sprm_['score'] = list_score
```

Count total sentiment

```python
temp = df_sprm_.groupby('predict').count()['tweet'].reset_index().sort_values(by='tweet',ascending=False)
temp.style.background_gradient(cmap='Purples')
```
